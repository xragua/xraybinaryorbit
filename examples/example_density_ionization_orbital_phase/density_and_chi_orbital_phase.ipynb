# Density and Ionization Parameter in the Orbital Path of Radiation

In this Jupyter notebook, we will provide an example of the usage of a function that provides the density and ionization parameter that the emited radiation trespasses in its path to the observer. All calculations assume that the wind is spherically symmetric, following the CAK model (Castor, Abbott, and Klein) as described in:

Castor, J. I., Abbott, D. C., & Klein, R. I. (1975). Radiation-driven winds in Of stars. *Astrophysical Journal*, 195, 157-174.

### Objectives
1. Calculate the wind density throughout the emitted radiation's path.
2. Calculate the ionization parameter through the emitted radiation's path.

### Methodology
- **Wind Density Calculation**:
  - Assume a spherically symmetric, unionized wind distribution.
  - Use the CAK model to describe the wind properties.

- **Geometrical Considerations**:
  - The result is depending on the orbital phase of the compact object (the radiation emission location).
  - The density depends on the distance to the donnor while the ionization parameter depends on the density and the distance to the compact object.
